```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a 
#heatmap


import pandas as pd
np.random.seed(42)

num_samples = 100
x1 = np.random.rand(num_samples)
x2 = x1 + np.random.normal(0, 0.1, num_samples) 
x3 = np.random.rand(num_samples)
x4 = 0.5 * x1 + 0.5 * x3 + np.random.normal(0, 0.1, num_samples)  

data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

correlation_matrix = data.corr()
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))


sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
                                                        #BOKEH ASSIGNMENT:

# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

from bokeh.plotting import figure, show, output_notebook
import numpy as np

output_notebook()


x_values = np.linspace(0, 10, 100)
y_values = np.sin(x_values)


p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")
p.line(x_values, y_values, line_width=2, color="blue")


show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the 
#markers based on the 'sizes' and 'colors' columns


from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
import numpy as np
import random


output_notebook()

num_points = 100
x_values = np.random.rand(num_points) * 10  
y_values = np.random.rand(num_points) * 10  
sizes = np.random.rand(num_points) * 20 + 5  
colors = ["#" + ''.join([random.choice('0123456789ABCDEF') for _ in range(6)]) for _ in range(num_points)]

source = ColumnDataSource(data=dict(x=x_values, y=y_values, sizes=sizes, colors=colors))

p = figure(title="Random Scatter Plot", x_axis_label="x", y_axis_label="y")
p.scatter(x='x', y='y', size='sizes', color='colors', fill_alpha=0.6, source=source)
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap


output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))


p = figure(x_range=fruits, title="Fruit Counts", x_axis_label="Fruit", y_axis_label="Count", toolbar_location=None, tools="")

p.vbar(x='fruits', top='counts', width=0.6, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=["#c9d9d3", "#718dbf", "#e84d60", "#ddb7b1"], factors=fruits))


p.legend.title = "Fruits"
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```

```python
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

output_notebook()
p = figure(title="Heatmap of Random Data", x_range=(0, 1), y_range=(0, 1),
           toolbar_location=None, tools="", x_axis_location="above")


color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())


p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)


color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')


show(p)
```
